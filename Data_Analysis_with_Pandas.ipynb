```python
df_clients = pd.read_csv('client_list.csv')
df_ids = pd.read_csv('customer_id_list.csv')

# df_clients içindeki sütun adlarını büyük harf formatına getiriyor. Böylece iki tablo birleşirken isimler uyumlu oluyor.
df_clients.rename(columns={'first_name': 'FIRST_NAME', 'gender': 'GENDER', 'Age': 'AGE'}, inplace=True)

# İki tabloyu ortak sütunlara göre birleştirir:
clients = pd.merge(df_clients, df_ids, on=['FIRST_NAME', 'LAST_NAME', 'GENDER', 'AGE'])
```

```python
df_sales = pd.read_csv('purchase_log.csv')

detailed_sales = pd.merge(clients, df_sales, on='CUSTOMER_ID', how='inner')
# Tabloları CUSTOMER_ID sütunu üzerinden eşleşenleri birleştirir.
# how='inner' Sadece iki tabloda da bulunan CUSTOMER_ID kayıtlarını alır.
```

```python
spenders = detailed_sales.groupby('CUSTOMER_ID')['PRICE'].sum() # Her müşteri için toplam harcama miktarı çıkar.
```

```python
top_client = spenders.idxmax() # En çok harcama yapan müşterinin ID’sini bulur.
top_amount = spenders.max() # O müşterinin harcadığı en yüksek toplam parayı bulur.
print(top_client, top_amount) # Müşterinin ID’sini ve toplam harcamasını ekrana yazar.
```

```python
detailed_sales[detailed_sales['CUSTOMER_ID'] == top_client] # En çok harcayan müşterinin bütün alışveriş geçmişi ekrana gelir.
```

```python
detailed_sales.to_csv('df_out.csv', index=False) # Dodyayı CSV dosyası olarak kaydeder (Dışarı aktarır). index=False, dosyaya satır numaralarını yazmaz.
```
